```python
# urban_mobility_pipeline.py
print("hi")
import os
import sys
from dotenv import load_dotenv
import warnings
from pyspark.sql import SparkSession, Window
from pyspark.sql.functions import col, count, countDistinct, mean, stddev, when, lit, lag, unix_timestamp, split


warnings.filterwarnings('ignore')

# Load environment variables
load_dotenv()

class Config:
    MONGO_URI = os.getenv('MONGO_URI')
    DATABASE_NAME = os.getenv('DATABASE_NAME')
    
    # File paths
    DATA_PATHS = {
        'delhi': 'data/delhi/',
        'bangalore': 'data/bangalore/'
    }
    
    @classmethod
    def setup_directories(cls):
        """Create necessary directories"""
        os.makedirs('outputs/visuals', exist_ok=True)
        os.makedirs('outputs/data', exist_ok=True)
        os.makedirs('models', exist_ok=True)
        os.makedirs('data/delhi', exist_ok=True)
        os.makedirs('data/bangalore', exist_ok=True)
        print("Directories created successfully!")

# Initialize configuration
Config.setup_directories()

import pandas as pd
import numpy as np
import pymongo
from pymongo import MongoClient
import glob
import json
from datetime import datetime

class SparkManager:
    """Manage Spark session with proper error handling"""
    
    @staticmethod
    def setup_spark():
        """Setup and return Spark session"""
        try:
            import findspark
            findspark.init()
            
            from pyspark.sql import SparkSession
            # from pyspark.sql.functions import *
            # from pyspark.sql.types import *
            # from pyspark.sql.window import Window
            
            # Create Spark session
            spark = SparkSession.builder \
                .appName("UrbanMobilityAnalysis") \
                .config("spark.sql.adaptive.enabled", "true") \
                .config("spark.sql.adaptive.coalescePartitions.enabled", "true") \
                .config("spark.driver.memory", "2g") \
                .config("spark.executor.memory", "2g") \
                .config("spark.sql.legacy.timeParserPolicy", "LEGACY") \
                .getOrCreate()
            
            spark.sparkContext.setLogLevel("WARN")
            print("Spark session created successfully!")
            return spark
            
        except ImportError as e:
            print(f" Spark dependencies not found: {e}")
            print("Please install: pip install pyspark findspark")
            return None
        except Exception as e:
            print(f"Error creating Spark session: {e}")
            return None

class DataLoader:
    """Handle GTFS file loading and validation"""
    
    @staticmethod
    def validate_gtfs_files(city):
        """Check if required GTFS files exist"""
        required_files = {
            'delhi': ['agency.txt', 'calendar.txt', 'routes.txt', 'stop_times.txt', 'stops.txt', 'trips.txt'],
            'bangalore': ['agency.txt', 'calendar.txt', 'routes.txt', 'stop_times.txt', 'stops.txt', 'trips.txt']
        }
        
        city_path = Config.DATA_PATHS[city]
        missing_files = []
        
        for file in required_files[city]:
            file_path = os.path.join(city_path, file)
            if not os.path.exists(file_path):
                missing_files.append(file)
            else:
                # Check if file has content
                file_size = os.path.getsize(file_path)
                if file_size == 0:
                    print(f"File is empty: {file}")
        
        if missing_files:
            print(f" Missing files for {city}: {missing_files}")
            return False
        else:
            print(f" All required files found for {city}")
            return True
    
    @staticmethod
    def load_gtfs_file(city, file_type):
        """Load a single GTFS file as pandas DataFrame"""
        file_path = os.path.join(Config.DATA_PATHS[city], f"{file_type}.txt")
        try:
            # Check file encoding and handle issues
            try:
                df = pd.read_csv(file_path, encoding='utf-8')
            except UnicodeDecodeError:
                df = pd.read_csv(file_path, encoding='latin-1')
            
            print(f"  Loaded {file_type}: {len(df)} records, {len(df.columns)} columns")
            
            # Show sample of data
            if len(df) > 0:
                print(f"     Columns: {list(df.columns)}")
                if len(df) < 5:
                    print(f"     Sample data:\n{df.head()}")
            
            return df
        except Exception as e:
            print(f"   Error loading {file_type}: {e}")
            return pd.DataFrame()

class MongoDBManager:
    """Manage MongoDB operations"""
    
    def __init__(self):
        self.config = Config()
        try:
            self.client = MongoClient(self.config.MONGO_URI, serverSelectionTimeoutMS=5000)
            # Test connection
            self.client.admin.command('ismaster')
            self.db = self.client[self.config.DATABASE_NAME]
            print(" MongoDB connected successfully!")
        except Exception as e:
            print(f" MongoDB connection failed: {e}")
            raise
    
    def import_city_data(self, city):
        """Import all GTFS data for a city to MongoDB"""
        if not DataLoader.validate_gtfs_files(city):
            return False
        
        print(f"\n Importing {city} data to MongoDB...")
        
        file_types = [f.replace('.txt', '') for f in os.listdir(Config.DATA_PATHS[city]) if f.endswith('.txt')]
        
        total_records = 0
        for file_type in file_types:
            df = DataLoader.load_gtfs_file(city, file_type)
            if not df.empty:
                collection_name = f"{city}_{file_type}"
                collection = self.db[collection_name]
                
                # Clear existing data
                collection.delete_many({})
                
                # Insert new data
                records = df.to_dict('records')
                if records:
                    collection.insert_many(records)
                    total_records += len(records)
                    
                    # Create indexes
                    self.create_indexes(collection, file_type)
        
        print(f" Imported {total_records} total records for {city}")
        return True
    
    def create_indexes(self, collection, file_type):
        """Create appropriate indexes for each collection type"""
        if file_type == 'routes':
            collection.create_index('route_id')
        elif file_type == 'trips':
            collection.create_index('trip_id')
            collection.create_index('route_id')
        elif file_type == 'stops':
            collection.create_index('stop_id')
        elif file_type == 'stop_times':
            collection.create_index('trip_id')
            collection.create_index('stop_id')
            collection.create_index([('trip_id', 1), ('stop_sequence', 1)])
    
    def get_database_stats(self):
        """Get statistics about the database"""
        stats = {}
        collections = self.db.list_collection_names()
        
        print("\n Database Collections:")
        for coll_name in collections:
            count = self.db[coll_name].count_documents({})
            stats[coll_name] = count
            print(f"  {coll_name}: {count} documents")
        
        return stats

class SparkProcessor:
    """Handle Spark data processing"""
    
    def __init__(self, spark):
        self.spark = spark
    
    def pandas_to_spark_with_schema(self, pandas_df):
        """Convert pandas DataFrame to Spark DataFrame with proper schema inference"""
        if pandas_df.empty:
            return None
        
        # Clean column names for Spark
        pandas_df.columns = [col.replace(' ', '_').replace('-', '_') for col in pandas_df.columns]
        
        try:
            return self.spark.createDataFrame(pandas_df)
        except Exception as e:
            print(f" Error converting to Spark DataFrame: {e}")
            # Fallback: create with string schema
            from pyspark.sql.types import StructType, StructField, StringType
            schema = StructType([StructField(col, StringType()) for col in pandas_df.columns])
            return self.spark.createDataFrame(pandas_df, schema)
    
    def process_city_data(self, city, mongo_manager):
        """Process GTFS data for a city using Spark"""
        print(f"\n Processing {city} data with Spark...")
        
        # Read data from MongoDB collections
        collections_data = {}
        for file_type in ['routes', 'trips', 'stop_times', 'stops']:
            collection_name = f"{city}_{file_type}"
            try:
                # Read from MongoDB using pandas
                collection = mongo_manager.db[collection_name]
                pandas_df = pd.DataFrame(list(collection.find({}, {'_id': 0})))
                
                if not pandas_df.empty:
                    spark_df = self.pandas_to_spark_with_schema(pandas_df)
                    collections_data[file_type] = spark_df
                    print(f"   Loaded {file_type}: {pandas_df.shape[0]} records")
                else:
                    print(f"    No data found for {file_type}")
                    collections_data[file_type] = None
                    
            except Exception as e:
                print(f"   Error loading {file_type}: {e}")
                collections_data[file_type] = None
        
        # Check if we have the essential data
        if not all(collections_data.values()):
            print(f" Missing essential data for {city}")
            return None
        
        # Process the data
        try:
            results = self.process_gtfs_data(collections_data, city)
            return results
        except Exception as e:
            print(f" Error processing {city} data: {e}")
            import traceback
            traceback.print_exc()
            return None
    
    def process_gtfs_data(self, collections_data, city):
        """Process GTFS data and calculate metrics"""
        routes = collections_data['routes']
        trips = collections_data['trips']
        stop_times = collections_data['stop_times']
        stops = collections_data['stops']
        
        # from pyspark.sql.functions import *
        # from pyspark.sql.window import Window
        
        print("   Cleaning and transforming data...")
        
        # Basic data cleaning
        routes_clean = routes.dropDuplicates(['route_id']).filter(col('route_id').isNotNull())
        trips_clean = trips.dropDuplicates(['trip_id']).filter(col('trip_id').isNotNull())
        stops_clean = stops.dropDuplicates(['stop_id']).filter(col('stop_id').isNotNull())
        
        # Clean stop_times - handle time formats and missing values
        stop_times_clean = stop_times \
            .filter(col('trip_id').isNotNull()) \
            .filter(col('stop_id').isNotNull()) \
            .filter(col('arrival_time').isNotNull())
        
        # Add time parsing with error handling
        stop_times_clean = stop_times_clean \
            .withColumn('arrival_time_clean', 
                       when(col('arrival_time').rlike('^\\d{1,2}:\\d{2}:\\d{2}$'), col('arrival_time'))
                       .otherwise(lit('00:00:00'))) \
            .withColumn('arrival_hour', 
                       split(col('arrival_time_clean'), ':').getItem(0).cast('int')) \
            .withColumn('arrival_minute',
                       split(col('arrival_time_clean'), ':').getItem(1).cast('int'))
        
        print("   Joining datasets...")
        
        # Create trip-stop events (main analysis table)
        trip_stop_events = stop_times_clean \
            .join(trips_clean, 'trip_id') \
            .join(routes_clean, 'route_id') \
            .join(stops_clean, 'stop_id')
        
        # Calculate basic route statistics
        print("   Calculating route statistics...")
        route_stats = trip_stop_events.groupBy('route_id').agg(
            countDistinct('trip_id').alias('num_trips'),
            countDistinct('stop_id').alias('num_stops'),
            count('*').alias('total_stop_events'),
            mean('arrival_hour').alias('avg_hour'),
            stddev('arrival_hour').alias('std_hour')
        )
        
        # Calculate headways (time between consecutive vehicles)
        print("   Calculating headways...")
        window_spec = Window.partitionBy('route_id', 'stop_id').orderBy('arrival_time_clean')
        
        headways = trip_stop_events \
            .withColumn('prev_arrival', lag('arrival_time_clean').over(window_spec)) \
            .withColumn('headway_seconds',
                       when(col('prev_arrival').isNotNull(),
                            (unix_timestamp(col('arrival_time_clean')) - 
                             unix_timestamp(col('prev_arrival'))))
                       .otherwise(lit(None))) \
            .filter(col('headway_seconds').isNotNull()) \
            .filter(col('headway_seconds') > 0) \
            .filter(col('headway_seconds') < 3600)  # Remove headways > 1 hour
        
        # Calculate hourly service frequency
        print("   Calculating service frequency...")
        hourly_service = trip_stop_events.groupBy('route_id', 'arrival_hour').agg(
            count('*').alias('hourly_stop_events'),
            countDistinct('trip_id').alias('hourly_trips')
        ).orderBy('route_id', 'arrival_hour')
        
        return {
            'trip_stop_events': trip_stop_events,
            'route_stats': route_stats,
            'headways': headways,
            'hourly_service': hourly_service
        }
    
    def save_spark_results(self, results, city):
        """Save Spark results to CSV and MongoDB"""
        if not results:
            return
        
        print(f"   Saving {city} results...")
        
        # Convert to pandas and save as CSV
        for result_name, spark_df in results.items():
            try:
                pandas_df = spark_df.limit(10000).toPandas()  # Limit for large datasets
                output_path = f"outputs/data/{city}_{result_name}.csv"
                pandas_df.to_csv(output_path, index=False)
                print(f"     CSV: {output_path} ({len(pandas_df)} records)")
                
                # Also save to MongoDB
                self.save_to_mongodb(pandas_df, f"{city}_analytics_{result_name}")
                
            except Exception as e:
                print(f"     Error saving {result_name}: {e}")
    
    def save_to_mongodb(self, pandas_df, collection_name):
        """Save DataFrame to MongoDB"""
        try:
            client = MongoClient(Config.MONGO_URI)
            db = client[Config.DATABASE_NAME]
            collection = db[collection_name]
            
            # Clear existing data
            collection.delete_many({})
            
            # Insert new data
            if not pandas_df.empty:
                records = pandas_df.to_dict('records')
                collection.insert_many(records)
                print(f"     MongoDB: {collection_name} ({len(records)} records)")
                
        except Exception as e:
            print(f"     MongoDB save error for {collection_name}: {e}")

class AnalysisEngine:
    """Perform data analysis and visualization"""
    
    def __init__(self):
        self.config = Config()
    
    def load_analysis_data(self, city):
        """Load processed data for analysis"""
        data_files = {
            'trip_stop_events': f'outputs/data/{city}_trip_stop_events.csv',
            'route_stats': f'outputs/data/{city}_route_stats.csv',
            'headways': f'outputs/data/{city}_headways.csv',
            'hourly_service': f'outputs/data/{city}_hourly_service.csv'
        }
        
        loaded_data = {}
        for name, file_path in data_files.items():
            if os.path.exists(file_path):
                try:
                    loaded_data[name] = pd.read_csv(file_path)
                    print(f" Loaded {name}: {len(loaded_data[name])} records")
                except Exception as e:
                    print(f"Error loading {name}: {e}")
            else:
                print(f"File not found: {file_path}")
        
        return loaded_data
    
    def perform_comprehensive_analysis(self, city):
        """Perform comprehensive analysis for a city"""
        print(f"\n Performing comprehensive analysis for {city}...")
        
        data = self.load_analysis_data(city)
        
        if not data:
            print(f" No analysis data found for {city}")
            return None
        
        analysis_results = {'city': city, 'timestamp': datetime.now().isoformat()}
        
        # Basic statistics
        if 'route_stats' in data:
            stats = data['route_stats']
            analysis_results['basic_stats'] = {
                'total_routes': len(stats),
                'avg_trips_per_route': stats['num_trips'].mean(),
                'avg_stops_per_route': stats['num_stops'].mean(),
                'total_stop_events': stats['total_stop_events'].sum(),
                'max_trips_route': stats['num_trips'].max(),
                'min_trips_route': stats['num_trips'].min()
            }
            
            print(f" Basic Statistics for {city}:")
            for key, value in analysis_results['basic_stats'].items():
                print(f"   {key}: {value:.2f}")
        
        # Headway analysis
        if 'headways' in data and 'headway_seconds' in data['headways'].columns:
            headways = data['headways']
            headways_min = headways['headway_seconds'] / 60
            
            analysis_results['headway_analysis'] = {
                'avg_headway_min': headways_min.mean(),
                'median_headway_min': headways_min.median(),
                'std_headway_min': headways_min.std(),
                'p95_headway_min': headways_min.quantile(0.95),
                'total_headways_analyzed': len(headways_min)
            }
            
            print(f" Headway Analysis for {city}:")
            for key, value in analysis_results['headway_analysis'].items():
                print(f"   {key}: {value:.2f}")
        
        # Service patterns
        if 'hourly_service' in data:
            hourly = data['hourly_service']
            analysis_results['service_patterns'] = {
                'peak_hour': hourly.groupby('arrival_hour')['hourly_trips'].sum().idxmax(),
                'off_peak_hour': hourly.groupby('arrival_hour')['hourly_trips'].sum().idxmin(),
                'total_service_hours': hourly['arrival_hour'].nunique()
            }
        
        return analysis_results
    
    def create_comprehensive_visualizations(self, city, data):
        """Create comprehensive visualizations"""
        try:
            import matplotlib.pyplot as plt
            import seaborn as sns
            
            print(f"Creating visualizations for {city}...")
            
            plt.style.use('default')
            sns.set_palette("husl")
            
            # 1. Route trips distribution
            if 'route_stats' in data:
                self._plot_route_distributions(city, data['route_stats'])
            
            # 2. Headway analysis
            if 'headways' in data and 'headway_seconds' in data['headways'].columns:
                self._plot_headway_analysis(city, data['headways'])
            
            # 3. Hourly service patterns
            if 'hourly_service' in data:
                self._plot_service_patterns(city, data['hourly_service'])
                
        except ImportError:
            print(" matplotlib/seaborn not available for visualizations")
        except Exception as e:
            print(f" Error creating visualizations: {e}")
    
    def _plot_route_distributions(self, city, route_stats):
        """Plot route distribution charts"""
        import matplotlib.pyplot as plt
        import seaborn as sns
        
        fig, axes = plt.subplots(2, 2, figsize=(15, 12))
        
        # Trips per route
        axes[0,0].hist(route_stats['num_trips'], bins=30, alpha=0.7, color='skyblue')
        axes[0,0].set_title(f'Trips per Route - {city.title()}')
        axes[0,0].set_xlabel('Number of Trips')
        axes[0,0].set_ylabel('Frequency')
        
        # Stops per route
        axes[0,1].hist(route_stats['num_stops'], bins=30, alpha=0.7, color='lightgreen')
        axes[0,1].set_title(f'Stops per Route - {city.title()}')
        axes[0,1].set_xlabel('Number of Stops')
        axes[0,1].set_ylabel('Frequency')
        
        # Top 10 busiest routes
        top_routes = route_stats.nlargest(10, 'num_trips')
        axes[1,0].bar(range(len(top_routes)), top_routes['num_trips'], color='coral')
        axes[1,0].set_title(f'Top 10 Busiest Routes - {city.title()}')
        axes[1,0].set_xlabel('Route Rank')
        axes[1,0].set_ylabel('Number of Trips')
        axes[1,0].set_xticks(range(len(top_routes)))
        axes[1,0].set_xticklabels(top_routes['route_id'], rotation=45)
        
        # Stop events distribution
        axes[1,1].hist(route_stats['total_stop_events'], bins=30, alpha=0.7, color='gold')
        axes[1,1].set_title(f'Stop Events per Route - {city.title()}')
        axes[1,1].set_xlabel('Total Stop Events')
        axes[1,1].set_ylabel('Frequency')
        
        plt.tight_layout()
        plt.savefig(f'outputs/visuals/{city}_route_analysis.png', dpi=300, bbox_inches='tight')
        plt.close()
        print(f"Saved route analysis: outputs/visuals/{city}_route_analysis.png")
    
    def _plot_headway_analysis(self, city, headways):
        """Plot headway analysis charts"""
        import matplotlib.pyplot as plt
        
        headways_min = headways['headway_seconds'] / 60
        
        fig, axes = plt.subplots(1, 2, figsize=(12, 5))
        
        # Headway distribution
        axes[0].hist(headways_min[headways_min < 60], bins=50, alpha=0.7, color='purple')
        axes[0].set_title(f'Headway Distribution - {city.title()}')
        axes[0].set_xlabel('Headway (minutes)')
        axes[0].set_ylabel('Frequency')
        
        # Headway box plot
        axes[1].boxplot(headways_min[headways_min < 60])
        axes[1].set_title(f'Headway Statistics - {city.title()}')
        axes[1].set_ylabel('Minutes')
        
        plt.tight_layout()
        plt.savefig(f'outputs/visuals/{city}_headway_analysis.png', dpi=300, bbox_inches='tight')
        plt.close()
        print(f" Saved headway analysis: outputs/visuals/{city}_headway_analysis.png")
    
    def _plot_service_patterns(self, city, hourly_service):
        """Plot service pattern charts"""
        import matplotlib.pyplot as plt
        
        hourly_agg = hourly_service.groupby('arrival_hour')['hourly_trips'].sum().reset_index()
        
        plt.figure(figsize=(10, 6))
        plt.plot(hourly_agg['arrival_hour'], hourly_agg['hourly_trips'], 
                marker='o', linewidth=2, markersize=6, color='teal')
        plt.title(f'Hourly Service Pattern - {city.title()}')
        plt.xlabel('Hour of Day')
        plt.ylabel('Number of Trips')
        plt.grid(True, alpha=0.3)
        plt.xticks(range(0, 24))
        plt.tight_layout()
        plt.savefig(f'outputs/visuals/{city}_service_patterns.png', dpi=300, bbox_inches='tight')
        plt.close()
        print(f" Saved service patterns: outputs/visuals/{city}_service_patterns.png")

def main():
    """Main execution function"""
    print(" Starting Urban Mobility Intelligence Pipeline")
    print("=" * 60)
    
    # Step 1: Setup Spark
    print("\n1.  Spark Setup")
    spark = SparkManager.setup_spark()
    
    if spark is None:
        print("Spark setup failed. Please install Spark dependencies.")
        print("Run: pip install pyspark findspark")
        return
    
    # Step 2: MongoDB Setup and Data Import
    print("\n2.  MongoDB Setup and Data Import")
    try:
        mongo_manager = MongoDBManager()
        
        for city in ['delhi', 'bangalore']:
            print(f"\n Processing {city.upper()}...")
            success = mongo_manager.import_city_data(city)
            if success:
                print(f" Successfully imported {city} data")
            else:
                print(f" Failed to import {city} data")
        
        # Show database statistics
        print("\n" + "="*50)
        mongo_manager.get_database_stats()
        
    except Exception as e:
        print(f" MongoDB operations failed: {e}")
        return
    
    # Step 3: Spark Processing
    print("\n3. Spark Data Processing")
    spark_processor = SparkProcessor(spark)
    
    all_analysis_results = []
    
    for city in ['delhi', 'bangalore']:
        print(f"\n Processing {city.upper()} with Spark...")
        results = spark_processor.process_city_data(city, mongo_manager)
        
        if results:
            spark_processor.save_spark_results(results, city)
            print(f" Successfully processed {city} data")
        else:
            print(f" Failed to process {city} data with Spark")
    
    # Step 4: Analysis and Visualization
    print("\n4. Data Analysis and Visualization")
    analysis_engine = AnalysisEngine()
    
    for city in ['delhi', 'bangalore']:
        print(f"\n Analyzing {city.upper()}...")
        analysis_results = analysis_engine.perform_comprehensive_analysis(city)
        
        if analysis_results:
            all_analysis_results.append(analysis_results)
            
            # Load data for visualization
            data = analysis_engine.load_analysis_data(city)
            if data:
                analysis_engine.create_comprehensive_visualizations(city, data)
    
    # Step 5: Generate Summary Report
    print("\n5. Generating Summary Report")
    generate_summary_report(all_analysis_results)
    
    # Clean up Spark
    if spark:
        spark.stop()
        print(" Spark session stopped")
    
    print("\n Pipeline completed successfully!")
    print(" Check the 'outputs/' folder for results and visualizations")

def generate_summary_report(analysis_results):
    """Generate a comprehensive summary report"""
    report = {
        "project": "Urban Mobility Intelligence System",
        "execution_time": datetime.now().isoformat(),
        "cities_analyzed": [],
        "summary": {},
        "key_findings": []
    }
    
    for result in analysis_results:
        city = result['city']
        report["cities_analyzed"].append(city)
        
        if 'basic_stats' in result:
            report["summary"][city] = result['basic_stats']
            
            # Add key findings
            findings = [
                f"{city.title()} has {result['basic_stats']['total_routes']} routes",
                f"Average {result['basic_stats']['avg_trips_per_route']:.1f} trips per route",
                f"Total {result['basic_stats']['total_stop_events']:,} stop events"
            ]
            
            if 'headway_analysis' in result:
                findings.append(f"Average headway: {result['headway_analysis']['avg_headway_min']:.1f} minutes")
            
            report["key_findings"].extend(findings)
    
    # Save report
    with open('outputs/data/comprehensive_analysis_report.json', 'w') as f:
        json.dump(report, f, indent=2)
    
    print("Comprehensive report generated: outputs/data/comprehensive_analysis_report.json")
    
    # Print executive summary
    print("\nEXECUTIVE SUMMARY")
    print("=" * 50)
    for finding in report["key_findings"]:
        print(f" {finding}")

if __name__ == "__main__":
    main()
```
